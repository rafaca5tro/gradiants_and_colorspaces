```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#Choose image (Frame example)
image = mpimg.imread('test4.jpg')
#image = mpimg.imread('test6.jpg')
thresh = (180, 255)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
binary = np.zeros_like(gray)
binary[(gray > thresh[0]) & (gray <= thresh[1])] = 1
```

```python
# Plot the result
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(gray,cmap = 'gray')
ax1.set_title('Gray', fontsize=35)
ax2.imshow(binary, cmap='gray')
ax2.set_title('Binary HLS', fontsize=35)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
R = image[:,:,0]
G = image[:,:,1]
B = image[:,:,2]
```

```python
# Plot the result
f, (ax3, ax4, ax5) = plt.subplots(1, 3,  figsize=(16, 6))
f.tight_layout()
ax3.imshow(R,cmap = 'gray')
ax3.set_title('R', fontsize=15)
ax4.imshow(G, cmap='gray')
ax4.set_title('G', fontsize=15)
ax5.imshow(B, cmap='gray')
ax5.set_title('B', fontsize=15)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
thresh = (200, 255)
binary_R = np.zeros_like(R)
binary_R[(R > thresh[0]) & (R <= thresh[1])] = 1
```

```python
# Plot the result
f, (ax6, ax7) = plt.subplots(1, 2,  figsize=(16, 6))
f.tight_layout()
ax6.imshow(R,cmap = 'gray')
ax6.set_title('R', fontsize=15)
ax7.imshow(binary_R, cmap='gray')
ax7.set_title('R Binary', fontsize=15)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
H = hls[:,:,0]
L = hls[:,:,1]
S = hls[:,:,2]
```

```python
# Plot the result
f, (ax8, ax9, ax10) = plt.subplots(1, 3,  figsize=(16, 6))
f.tight_layout()
ax8.imshow(H,cmap = 'gray')
ax8.set_title('H', fontsize=15)
ax9.imshow(L, cmap='gray')
ax9.set_title('L', fontsize=15)
ax10.imshow(S, cmap='gray')
ax10.set_title('S', fontsize=15)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
thresh = (90, 255)
binary_S = np.zeros_like(S)
binary_S[(S > thresh[0]) & (S <= thresh[1])] = 1
```

```python
# Plot the result
f, (ax11, ax12) = plt.subplots(1, 2,  figsize=(16, 6))
f.tight_layout()
ax11.imshow(S,cmap = 'gray')
ax11.set_title('S', fontsize=15)
ax12.imshow(binary_S, cmap='gray')
ax12.set_title('S Binary', fontsize=15)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```
